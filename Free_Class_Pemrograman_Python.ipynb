# Gabung, Intersect, Difference & More

Saatnya eksplor operasi lanjutan di `set`. Dijamin seru!

### Union (Gabung)
Balikin `set` baru yang isinya gabungan dua `set`, tanpa duplikat:
```python
a = {"a","b","c"}
b = {1,2,3}
c = a.union(b)
# atau bisa pakai: c = a | b
```

### Intersection (Irisan)

Cuma ambil elemen yang ada di dua `set`:
```python
c = a.intersection(b)
# atau bisa pakai: c = a & b
```

### Difference (Selisih)

Elemen yang cuma ada di `set` satu aja, gak di `set` kedua:

```python
c = a.difference(b)
# atau bisa pakai: c = a - b
```

### Symmetric Difference

Elemen yang cuma ada di salah satu `set`, tapi bukan di keduanya:
```python
c = a.symmetric_difference(b)
# atau pakai: c = a ^ b
```

# Kenalan Sama While Loop di Python

Sekarang waktunya belajar **while loop** di Python‚Äîcara seru untuk mengulang kode selama kondisi masih *True*!

### Struktur Dasarnya:
```python
i = 1
while i < 6:
    print(i)
    i += 1
```
**Penjelasan:**

- `i` dimulai dari 1.
- Selama `i < 6` masih `True`, kode di dalam while terus jalan.
- Jangan lupa naikkan `i` (increment) supaya loop-nya gak **infinite loop**!

# `break` & `continue` di While Loop

###  `break`
Kalau mau berhenti sebelum kondisi terpenuhi:
```python
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1
```

Loop berhenti begitu `i` jadi 3.

### `continue`

Lewati iterasi yang tidak diinginkan:

```python
i = 0
while i < 6:
    i += 1
    if i == 3:
        continue
    print(i)
```

Waktu i == 3, loop langsung lanjut ke iterasi berikutnya tanpa print
